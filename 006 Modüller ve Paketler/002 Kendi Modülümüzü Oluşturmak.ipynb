### İlk Modülünüz: Adım 2

İkinci dosya ise, yeni modülü kullanacak olan `main.py` dosyasıdır. Bu dosyanın içeriği şu an için oldukça kısa:

#### import module Talimatını İçeren main.py Dosyasını Oluşturma

```python

```

#### Not:
Her iki dosyanın da aynı klasörde bulunması gerekiyor. Her iki dosya için yeni, boş bir klasör oluşturmanızı şiddetle tavsiye ederiz. Bu, işleri daha kolay hale getirecektir.

IDLE'ı (veya tercih ettiğiniz başka bir IDE'yi) açın ve `main.py` dosyasını çalıştırın. Ne görüyorsunuz?

Hiçbir şey görmemeniz gerekiyor. Bu, Python'ın `module.py` dosyasının içeriğini başarıyla içe aktardığı anlamına gelir. Şu anda modül boş olsa da, bu ilk adımdır. Bir sonraki adıma geçmeden önce, her iki dosyanın bulunduğu klasöre bir göz atın.

İlginç bir şey fark ediyor musunuz?

Yeni bir alt klasör ortaya çıktı - görebiliyor musunuz? Adı `__pycache__`. İçine bir bakın. Ne görüyorsunuz?

İçerisinde, adında (aşağı yukarı) `module.cpython-xy.pyc` olan bir dosya var, burada `x` ve `y` Python sürümünüzden türetilen rakamlardır (örneğin, Python 3.8 kullanıyorsanız 3 ve 8 olacak).

Dosyanın adı, modülünüzün adıyla aynıdır (burada `module`). İlk noktadan sonraki kısım, dosyanın hangi Python implementasyonu tarafından oluşturulduğunu (burada CPython) ve sürüm numarasını belirtir. Son kısım (pyc), Python ve derlenmiş kelimelerinden gelir.

Dosyanın içeriğine bakabilirsiniz, ancak insanlar için tamamen okunamaz olacaktır. Bu normaldir, çünkü dosya yalnızca Python'un kullanımı içindir.

Python, bir modülü ilk kez içe aktardığında, içeriğini yarı derlenmiş bir şekle çevirir. Dosya makine kodu içermez, ancak Python'un yorumlayıcısı tarafından çalıştırılmaya hazır, içsel Python yarı derlenmiş kodu içerir. Bu tür bir dosya, saf kaynak dosyası için gereken birçok kontrolü gerektirmez, bu yüzden çalıştırma daha hızlı başlar ve daha hızlı çalışır.

Bu sayede, sonraki her içe aktarma işlemi, kaynak metni baştan yorumlamaktan daha hızlı olacaktır.

Python, modülün kaynak dosyasının değiştirilip değiştirilmediğini kontrol edebilir. Eğer değiştirilmişse, pyc dosyası yeniden oluşturulacaktır. Değiştirilmemişse, pyc dosyası hemen çalıştırılabilir. Bu süreç tamamen otomatik ve şeffaftır, bu yüzden aklınızda tutmanıza gerek yoktur.

### İlk Modülünüz: Adım 4

#### import module Talimatını İçeren main.py Dosyası

```python
import module
```

`main.py` dosyasını çalıştırın. Şu satırı görmelisiniz:

```
Ben bir modül olmayı seviyorum.
```

Bu, bir modül içe aktarıldığında içeriğinin Python tarafından çalıştırıldığı anlamına gelir. Bu, modülün bazı iç bileşenlerini (örneğin, bazı değişkenlere değerler atamak) başlatmasına olanak tanır.

Şu senaryoyu düşünün:

- `mod1` adında bir modül var.
- `mod1` modülünü içe aktaran `mod2` adında başka bir modül var.
- Hem `mod1` hem de `mod2` modüllerini içe aktaran bir `main.py` dosyası var.

İlk bakışta, `mod1` modülünün iki kez içe aktarılacağını düşünebilirsiniz, ancak Python yalnızca ilk içe aktarmayı gerçekleştirir. Python, içe aktarılan modülleri hatırlar ve sonraki içe aktarmaları sessizce atlar.

### İlk Modülünüz: Adım 5

Python ayrıca `__name__` adlı özel bir değişken oluşturur. Her kaynak dosyasının bu değişkenin kendi versiyonu vardır. Bunu göstermek için modülü biraz değiştirelim:

#### module.py Dosyasını Güncelleme

```python
print("Ben bir modül olmayı seviyorum.")
print(__name__)
```

Şimdi `module.py` dosyasını çalıştırın. Şu satırları görmelisiniz:

```
Ben bir modül olmayı seviyorum
__main__
```

Sonra `main.py` dosyasını çalıştırın. Şu satırları görmelisiniz:

```
Ben bir modül olmayı seviyorum
module
```

Bu, bir dosya doğrudan çalıştırıldığında, `__name__` değişkeninin `__main__` olarak ayarlandığını gösterir. Dosya bir modül olarak içe aktarıldığında ise, `__name__` değişkeni dosyanın adına (dosya uzantısı .py olmadan) ayarlanır.

### İlk Modülünüz: Adım 6

### İlk Modülünüz: Adım 9

`module.py` dosyamıza yeni kodlar yazalım. İşte güncellenmiş modül:

#### module.py

```python
#!/usr/bin/env python3 

""" module.py - bir Python modülü örneği """

__counter = 0

def suml(the_list):
    global __counter
    __counter += 1
    the_sum = 0
    for element in the_list:
        the_sum += element
    return the_sum

def prodl(the_list):
    global __counter    
    __counter += 1
    prod = 1
    for element in the_list:
        prod *= element
    return prod

if __name__ == "__main__":
    print("Ben bir modül olmayı tercih ederim, ama sizin için bazı testler yapabilirim.")
    my_list = [i + 1 for i in range(5)]
    print(suml(my_list) == 15)
    print(prodl(my_list) == 120)
```

İşte birkaç açıklama:

- `#!` ile başlayan satır shebang olarak bilinir. Unix benzeri işletim sistemlerine dosyanın nasıl çalıştırılacağını söylerimportçin bu sadece bir yorum satırıdır.
- Modül talimatlarından (ithalatlar gibi) önce gelen bir dize (belki çok satırlı) doc-string olarak adlandırılır ve modülün amacı ve içeriği hakkında kısa bir açıklama yapar.
- Modül içinde tanımlanan fonksiyonlar (`suml()` ve `prodl()`) içe aktarılabilir.
- Dosya doğrudan çalıştırıldığında `__name__` değişkenini algılamak için kullanılır, bu da bize bazı basit testler yapma fırsatı verir.


### İlk Modülünüz: Adım 11

Artık örneğimizi biraz daha karmaşık hale getirelim. Şimdiye kadar ana Python dosyasının ve içe aktarılan modülün aynı klasörde olduğunu varsaydık. Bu varsayımı değiştirelim ve aşağıdaki senaryoyu düşünelim:

- Windows® işletim sistemi kullanıyoruz (dosya yolunun biçimi buna bağlı olarak değişir).
- Ana Python betiği `C:\Users\user\py\progs` klasöründe bulunuyor ve adı `main.py`.
- İçe aktarılacak modül `C:\Users\user\py\modules` klasöründe bulunuyor.

### Python Modüllerini Arama Yöntemi

Bu senaryoyu yönetmek için Python'un modüllerini nasıl aradığını anlamamız gerekiyor. Python, istemci tarafından istenen bir modülü bulmak için kullanılan tüm konumları (klasörleri) içeren özel bir değişken (aslında bir liste) kullanır.

Python, bu klasörleri listede belirtilen sırayla tarar. Modülü bu klasörlerden birinde bulamazsa, içe aktarma başarısız olur. Aranan isimle eşleşen ilk klasör dikkate alınır ve geriye kalan klasörlerde bu isimde başka bir modül varsa göz ardı edilir.

Bu değişken `path` olarak adlandırılır ve `sys` modülü aracılığıyla erişilebilir. Geçerli değerini kontrol etmek için şu şekilde yapabilirsiniz:

```python
import sys

for p in sys.path:
    print(p)
```

Bu kodu `C:\Users\user` dizininden çalıştırdığımızda, benzer bir çıktı alırız:

```plaintext
C:\Users\user
C:\Users\user\AppData\Local\Programs\Python\Python36-32\python36.zip
C:\Users\user\AppData\Local\Programs\Python\Python36-32\DLLs
C:\Users\user\AppData\Local\Programs\Python\Python36-32\lib
C:\Users\user\AppData\Local\Programs\Python\Python36-32
C:\Users\user\AppData\Local\Programs\Python\Python36-32\lib\site-packages
```

Not: `path` listesindeki ilk öğe, yürütmenin başladığı dizindir.

Ayrıca, zip dosyasının bir yol öğesi olarak listelendiğini görebilirsiniz. Bu bir hata değildir. Python zip dosyalarını normal klasörler gibi işleyebilir, bu da depolama alanı kazandırabilir.

### Sorunu Çözme

Sorunumuzu çözmek için, modülü içeren klasörü `path` değişkenine ekleyebiliriz (bu değiştirilebilir).

### İlk Modülünüz: Adım 12

Muhtemel çözümlerden biri şu şekildedir:

#### `main.py` Dosyasını Güncelleme

```python
from sys import path

path.append('..\\modules')

import module

zeroes = [0 for i in range(5)]
ones = [1 for i in range(5)]
print(module.suml(zeroes))
print(module.prodl(ones))
```

### Önemli Notlar

1. **Ters Eğik Çizgilerin İki Kez Kullanılması:** Klasör adındaki `\` işaretini iki kez kullandık. Bu gereklidir çünkü Python dizelerinde tek bir ters eğik çizgi bir kaçış karakteridir.

2. **Göreceli Klasör Adı:** Klasör için göreceli bir yol kullandık. Bu, `main.py` dosyasını doğrudan başlatırsanız çalışır ve mevcut dizin göreceli yol ile eşleşmiyorsa çalışmaz. Mevcut dizin eşleşmiyorsa her zaman mutlak bir yol kullanabilirsiniz, örneğin:

    ```python
    path.append('C:\\Users\\user\\py\\modules')
    ```

3. **`append()` Kullanımı:** `append()` yöntemini kullandık, bu yeni yolu `path` listesinin son öğesi olarak ekler. Eğer fikrinizi beğenmezseniz, bunun yerine `insert()` kullanabilirsiniz.efer to place it elsewhere in the list, you can use `insert()` instead.
