### Windows'ta pip Kurulumu

1. **Python Kurulumu**: Öncelikle, sisteminizde Python yüklü olmalıdır. [Python web sitesinden](https://www.python.org/downloads/) Python yükleyicisini indirip çalıştırın. Kurulum sırasında "Add Python to PATH" seçeneğini işaretlemeyi unutmayın.

2. **get-pip.py İndir**: Web tarayıcınızı açın ve [get-pip.py](https://bootstrap.pypa.io/get-pip.py) adresine gidin. Sayfaya sağ tıklayıp "Farklı Kaydet" seçeneği ile dosyayı bilgisayarınıza indirin.

3. **get-pip.py Çalıştır**: Komut İstemi'ni (cmd) açın. `cd` komutunu kullanarak get-pip.py dosyasını indirdiğiniz dizine gidin. Aşağıdaki komutu yazarak script'i çalıştırın:
   ```sh
   python get-pip.py
   ```

4. **Kurulumu Doğrula**: Kurulum tamamlandıktan sonra, pip'in doğru kurulduğunu doğrulamak için aşağıdaki komutu yazın:
   ```sh
   pip --version
   ```

Bu komut, yüklü olan pip sürümünü göstererek pip'in başarılı bir şekilde kurulduğunu teyit edecektir.

### Linux'ta pip Kurulumu

Linux sisteminde pip kurmak için aşağıdaki adımları izleyin:

1. **Paket Listesini Güncelle**:
   Terminali açın ve paket listesini güncelleyin:
   ```sh
   sudo apt update
   ```

2. **Python 3 için pip Kur**:
   Python 3 için pip'i aşağıdaki komutla kurun:
   ```sh
   sudo apt install python3-pip
   ```

3. **Kurulumu Doğrula**:
   Pip'in kurulumunu doğrulamak için sürümünü kontrol edin:
   ```sh
   pip3 --version
   ```

Bu adımlarla pip'i Linux sisteminize kurabilir ve Python paketlerini yönetmek için kullanabilirsiniz.


Bazı Linux dağıtımları, Python yüklü olsa bile pip'i önceden yüklü olarak sunmaz (bazı Ubuntu sürümleri bu şekilde davranabilir). Bu durumda iki seçeneğiniz var:

1. pip'i özel bir paket yöneticisi kullanarak sistem paketi olarak kurmak (Debian tabanlı sistemlerde apt gibi).
2. pip'i Python’un dahili mekanizmalarını kullanarak kurmak.

İlk seçenek daha iyidir. OS'yi atlayarak pip'i indiren ve kuran akıllı scriptler olsa da, bunları kullanmanızı önermiyoruz, çünkü sorun çıkarabilirler.

Örneğin, Ubuntu Budgie kullanırken pip3'ü başlatmaya çalıştık ve başarısız olduk. OS, python3-pip adlı paketi kurmak için apt kullanmamızı önerdi:

```sh
sudo apt install python3-pip
```

Bu iyi bir tavsiye, ancak bunun için yönetici haklarına ihtiyacınız olacak. Farklı Linux dağıtımları farklı paket yöneticileri kullanabilir (örneğin, Arch Linux için pacman, Red Hat türevleri için yum).

Bu öneriyi takip ettiğimizde OS, pip ile birlikte gerekli ek bileşenleri de kurdu. Bu normaldir ve endişelenmenize gerek yok.
