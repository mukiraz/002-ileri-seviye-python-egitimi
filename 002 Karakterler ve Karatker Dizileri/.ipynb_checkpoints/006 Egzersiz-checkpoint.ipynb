**Tahmini Süre:**
30 dakika

**Zorluk Seviyesi:**
Orta

**Amaçlar:**
- Öğrencinin stringlerle çalışma becerilerini geliştirmek
- Stringleri metin dışı verileri temsil etmek için kullanmak

**Senaryo:**
Yedi segmentli bir ekranı mutlaka görmüşsünüzdür. Bu, yedi segmentin bir alt kümesini kullanarak bir ondalık basamağı göstermek için tasarlanmış bir cihazdır (bazen elektronik, bazen mekanik). Eğer hala ne olduğunu bilmiyorsanız, şu [Wikipedia makalesine](https://en.wikipedia.org/wiki/Seven-segment_display) başvurabilirsiniz.

Göreviniz, yedi segmentli bir ekranın çalışma prensibini simüle edebilen bir program yazmaktır, ancak segmentler yerine tek LED'ler kullanacaksınız.

Her bir rakam, 13 LED'den oluşur (bazıları yanar, bazıları karanlık). İşte bunu şöyle hayal ediyoruz:

```
  # ### ### # # ### ### ### ### ### ### 
  #   #   # # # #   #     # # # # # # # 
  # ### ### ### ### ###   # ### ### # # 
  # #     #   #   # # #   # # #   # # # 
  # ### ###   # ### ###   # ### ### ###
```

Not: 8 sayısı tüm LED ışıklarının yandığını gösterir.

Kodunuz, kullanıcının girdiği herhangi bir negatif olmayan tam sayıyı göstermelidir.

**İpucu:** Tüm on rakamın desenlerini içeren bir liste kullanmak çok yararlı olabilir.
