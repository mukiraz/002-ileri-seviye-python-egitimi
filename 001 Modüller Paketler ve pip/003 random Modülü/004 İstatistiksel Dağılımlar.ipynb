# İstatistiksel Dağılımlara göre rastgele değer seçme

İstatistikte, bir **istatistiksel dağılım**, bir değişkenin olasılık dağılımıdır. Bir değişkenin olasılık dağılımı, değişkenin alabileceği değerleri ve bu değerlerin olasılıklarını tanımlar.

İstatistiksel dağılımların birçok farklı türü vardır. Bazı yaygın istatistiksel dağılımlar şunlardır:

* **Normal dağılım:** Normal dağılım, en yaygın istatistiksel dağılımdır. Normal dağılım, bir çan eğrisi şeklindedir ve ortalaması ve standart sapması ile tanımlanır.
* **Poisson dağılımı:** Poisson dağılımı, belirli bir süre içinde belirli bir olayın meydana gelme olasılığını tanımlar. Örneğin, bir saat içinde bir telefon hattına gelen telefon çağrılarının sayısının olasılığını tanımlamak için Poisson dağılımı kullanılabilir.
* **Binomial dağılım:** Binomial dağılım, belirli bir sayıda deneyde belirli bir sonucun elde edilme olasılığını tanımlar. Örneğin, bir madeni paranın 10 kez atılmasında 5 kez tura gelme olasılığını tanımlamak için binomial dağılımı kullanılabilir.
* **Hipergeometrik dağılım:** Hipergeometrik dağılım, belirli bir sayıda deneyde belirli bir sonucun elde edilme olasılığını tanımlar. Örneğin, bir zar 10 kez atıldığında 6'nın 3 kez gelme olasılığını tanımlamak için hipergeometrik dağılım kullanılabilir.

İstatistiksel dağılımların birçok farklı uygulaması vardır. İstatistiksel dağılımlar, istatistiksel analizlerde, olasılık teorisinde ve makine öğrenimde yaygın olarak kullanılır.

**İstatistiksel dağılımların bazı temel özellikleri şunlardır:**

* **Ortalama:** Bir değişkenin olasılık dağılımının ortalaması, değişkenin alabileceği değerlerin toplamının olasılıklarına göre ağırlıklı ortalamasıdır.
* **Standart sapma:** Bir değişkenin olasılık dağılımının standart sapması, değişkenin ortalamadan sapmalarının ortalamasıdır.
* **Mod:** Bir değişkenin olasılık dağılımının modu, değişkenin en sık aldığı değerdir.
* **Medyan:** Bir değişkenin olasılık dağılımının medyanı, değişkenin değerlerini yarıya bölen değerdir.

İstatistiksel dağılımların özellikleri, değişkenin olasılık dağılımının şeklini ve davranışını tanımlamaya yardımcı olur.

## uniform() metodu

uniform dağılıma uygun rastgele değer döndürür
Python'da random.uniform() metodu, belirli bir aralıkta rastgele bir kayan noktalı sayı oluşturmak için kullanılır. Metodun ilk parametresi, aralığın başlangıç değerini, ikinci parametresi ise aralığın bitiş değerini temsil eder.

```python
import random

# 0 ile 100 arasında rastgele bir kayan noktalı sayı oluştur
random_number = random.uniform(0, 100)
print(random_number)
```

## gammavariate() metodu

Python'da random.gammavariate() metodu, gamma dağılımına göre rastgele bir sayı üretmek için kullanılır. Gamma dağılımı, belirli bir süre içinde belirli bir olayın meydana gelme olasılığını tanımlayan bir istatistiksel dağılımdır.

random.gammavariate() metodunun iki parametresi vardır:

alpha: Gamma dağılımının şekil parametresi.
beta: Gamma dağılımının ölçek parametresi.

```python
import random

# alpha = 2, beta = 3 olan gamma dağılımına göre rastgele bir sayı üret
random_number = random.gammavariate(2, 3)
print(random_number)
```
