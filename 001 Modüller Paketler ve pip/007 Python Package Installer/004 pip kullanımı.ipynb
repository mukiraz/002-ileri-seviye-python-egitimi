### pip show

`pip list` komutu çok bilgi vermez. Neyse ki, yüklü herhangi bir paket hakkında daha fazla bilgi almanızı sağlayan bir komut var:

```sh
pip show package_name
```

Örneğin, pip hakkında bilgi almak için:

```sh
pip show pip
```

Bu komut, paketin meta verilerinden bilgi çıkarır. Çıktı, bağımlılıkları (`Requires:`) ve ona bağımlı olan paketleri (`Required-by:`) içerir; her ikisi delarınızı spesifik tutun.

````python
!pip show pandas
````

### pip ile Paket Kurulumu

Bir paketi bilgisayarınıza kurmak için pip kullanabilirsiniz. İki senaryo vardır:

1. **Kullanıcıya Özel Kurulum**: Bu, paketi sadece kendi kullanıcı hesabınıza kurar. Yönetici haklarına sahip değilseniz veya yerel bir kurulum tercih ediyorsanız bu yöntemi kullanın.
2. **Sistem Genelinde Kurulum**: Bu, paketi sistemdeki tüm kullanıcılar için kurar ve yönetici hakları gerektirir.

Yerel kurulum için `--user` seçeneğini kullanın:

```sh
pip install --user paket_adi
```

Bu seçeneği eklemezseniz, pip sistem genelinde kurulum yapar:

```sh
pip install paket_adi
```

Örneğin, `pygame` kurmak için:

- **Sistem Genelinde**:
  ```sh
  pip install pygame
  ```

- **Kullanıcıya Özel**:
  ```sh
  pip install --user pygame
  ```

Kurulumdan sonra, kurulu paketi doğrulamak ve daha fazla bilgi almak için `pip show paket_adi` ve `pip list` komutlarını kullanabilirsiniz.

### `pip install` Komutunun Ekstra Özellikleri

`pip install` komutunun iki önemli ek özelliği vardır:

1. **Bir Paketi Güncelleme**: Belirli bir paketin en son sürümünü kullandığınızdan emin olmak için:
   ```sh
   pip install -U paket_adi
   ```
   Burada `-U`, güncelleme anlamına gelir. Aynı amaçla `--user` seçeneğini de kullanabilirsiniz.

2. **Belirli Bir Sürümü Kurma**: Belirli bir sürümü kurmak için:
   ```sh
   pip install paket_adi==paket_sürümü
   ```
   Örneğin:
   ```sh
   pip install pygame==1.9.2
   ```
