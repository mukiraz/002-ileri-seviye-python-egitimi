### Paket Nedir?

Kendi modüllerinizi yazmak, sıradan betikler yazmaktan çok farklı değildir. Dikkat etmeniz gereken bazı özel noktalar vardır, ancak bunlar kesinlikle çok zor değildir. Bunu kısa sürede göreceksiniz


.

#### Paket-Modül-Fonksiyon Konsepti

İşte hatırlamanız gereken bazı önemli noktalar:

- Bir modül, fonksiyonlar için bir kap gibidir. Bir modüle birçok fonksiyonu sığdırabilir ve bunları dünya çapında paylaşabilirsiniz.
- Farklı uygulama alanlarındaki fonksiyonları aynı modülde karıştırmamak genellikle iyi bir fikirdir. Tıpkı bir kütüphanede bilimsel eserlerin çizgi romanlarla karıştırılmaması gerektiği gibi. Bu yüzden, fonksiyonlarınızı mantıklı bir şekilde gruplandırın ve modülü açık ve sezgisel bir şekilde adlandırın (örneğin, sabit disk bölümlendirme ve formatlama fonksiyonları içeren bir modüle "arcade_games" ismini vermeyin).
- Birçok modül oluşturmak biraz karmaşa yaratabilir. Eninde sonunda, fonksiyonlarınızı gruplandırdığınız gibi modüllerinizi de gruplandırmak isteyeceksiniz. Modülden daha genel bir kap var mı? Evet, var - buna paket denir. Modüller dünyasında, bir paket, dosyalar dünyasındaki bir klasör/dizinle benzer bir rol oynar.
