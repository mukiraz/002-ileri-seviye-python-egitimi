**Ayrı modüllerde bulunan fonksiyonlar**

```python
#! /usr/bin/env python3

""" modül: alpha """

def funA():
    return "Alpha"

if __name__ == "__main__":
    print("Ben bir modül olmayı tercih ederim.")
```

### İlk Paketiniz: Adım 3

**Modüller arasındaki ilişki - ağaç yapısı**

Bu yapı neredeyse bir paket (Python anlamında). Tam işlevsel ve operasyonel olması için ince detayları eksik. Birazdan tamamlayacağız.

`extra`'nın yeni oluşturulan paketin adı olduğunu varsayarsak (paketin kökü olarak düşünün), bu isimlendirme kuralı, ağaçtaki her varlığı açıkça adlandırmanıza olanak tanır.

Örneğin:

- `tau` paketindeki `funT()` adlı bir fonksiyonun konumu şöyle tanımlanabilir:

  ```python

  ```

- `extra` paketinin `ugly` alt paketinde bulunan `psi` modülünden gelen bir `funP()` fonksiyonu şöyle tanımlanabilir:

  ```python
  extra.ugly.psi.funP()
  ```

### İlk Paketiniz: Adım 4

Cevaplanması gereken iki önemli soru var:

1. Bu tür bir ağacı (aslında, bir alt ağacı) gerçek bir Python paketine nasıl dönüştürürsünüz? Başka bir deyişle, Python'a bu ağacın sadece bir grup gereksiz dosya değil, bir dizi modül olduğunu nasıl kanıtlarsınız?
2. Bu alt ağacı Python'a erişilebilir hale getirmek için nereye yerleştirirsiniz?

İlk sorunun cevabı şaşırtıcı: paketler de modüller gibi başlatma gerektirebilir.

Bir modülün başlatılması, modül dosyasının içindeki serbest kod (herhangi bir fonksiyonun parçası olmayan kod) tarafından yapılır. Bir paket bir dosya olmadığı için, bu teknik paketleri başlatmak için işe yaramaz.

Bunun yerine farklı bir yönteme ihtiyacınız var - Python, paketin klasöründe çok özel bir isimle bir dosya bulunmasını bekler: `__init__.py`.

Bu dosyanın içeriği, paketin herhangi bir modülü içe aktarıldığında yürütülür. Eğer özel bir başlatma işlemi yapmak istemiyorsanız, dosyayı boş bırakabilirsiniz, ancak mutlaka dahil etmelisiniz.

### İlk Paketiniz: Adım 5

Unutmayın: `__init__.py` dosyasının varlığı, nihayetinde paketi oluştururak devam edeceksiniz.

**Modüller arasındaki ilişki ve `__init__.py` dosyasının varlığı - ağaç yapısı**

Not: Sadece kök klasör değil, herhangi bir alt klasöre (alt paketlere) de `__init__.py` dosyasını koyabilirsiniz. Bu, bazı alt paketlerin özel işlem ve başlatma türleri gerektirdiğinde yararlı olabilir.

Şimdi ikinci soruyu cevaplama zamanı - cevap basit: herhangi bir yere. Tek yapmanız gereken, Python'ın paketin yerini bilmesini sağlamak. Bunu nasıl yapacağınızı zaten biliyorsunuz.

Artık ilk paketinizi kullanmaya hazırsınız.

### İlk Paketiniz: Adım 7

`extra` paketinin en üstünden başlayarak `iota` modülündeki `funI()` fonksiyonuna erişeceğiz. Bu, tam nitelikli paket adlarını kullanmamızı gerektirir (bu, klasör ve alt klasörlerin adlandırılmasıyla çok benzerdir).

İşte nasıl yapılacağı:

**main2.py dosyası**

```python
from sys import path
path.append('..\\packages')

import extra.iota
print(extra.iota.funI())
```

**main2.py**

Not:

- Paketin Python tarafından erişilebilir olmasını sağlamak için `path` değişkenini değiştirdik.
- İthalat doğrudan modüle işaret etmiyor, ancak paketin en üstünden tam nitelikli yolu belirtiyor.
- `import extra.iota` yerine `import iota` kullanmak bir hataya neden olacaktır.

Aşağıdaki alternatif sürüm de geçerlidir:

**main2.py dosyasının alternatif sürümü**

```python
from sys import path
path.append('..\\packages')

from extra.iota import funI
print(funI())
```

**main2.py**

`iota` modülünün tam nitelikli adına dikkat edin.

### İlk Paketiniz: Adım 9

Şimdi varsayalım ki `extra` klasöründen başlayarak tüm alt dizini sıkıştırdık (içindeki her şeyi), ve `extrapack.zip` adında bir dosya elde ettik. Sonrasında bu dosyayı `packages` klasörüne koyduk.

Artık zip dosyasını bir paket olarak kullanabiliriz:

```python
from sys import path

path.append('..\\packages\\extrapack.zip')

import extra.good.best.sigma as sig
import extra.good.alpha as alp
from extra.iota import funI
from extra.good.beta import funB

print(sig.funS())
print(alp.funA())
print(funI())
print(funB())
```
